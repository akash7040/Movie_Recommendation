**Movie Recommendation System**

Recommender System:- A recommender system, also known as a recommendation system or engine, is a type of software application or algorithm designed to provide personalized suggestions or recommendations to users. These recommendations are typically based on the user's preferences, behaviors, or past interactions with the system. The goal of a recommender system is to help users discover items (such as products, movies, music, books, articles, etc.) that they might find interesting or relevant.

```python
df_features=df[['Movie_Genre','Movie_Keywords','Movie_Tagline', 'Movie_Cast','Movie_Director']].fillna('')
```

```python
x=df_features['Movie_Genre']+' '+df_features['Movie_Keywords']+' '+df_features['Movie_Tagline']+' '+df_features['Movie_Cast']+' '+df_features['Movie_Director']
```

```python
x
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer()
```

```python
x=tfidf.fit_transform(x)
```

```python
print(x)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_score=cosine_similarity(x)
```
